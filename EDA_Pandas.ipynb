```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv("/content/drive/MyDrive/lecture_data/csv_data/titanic_data.csv")
```

```python
titanic_data.head(10)
```

```python
unique_values = titanic_data['Pclass'].unique()

print(unique_values)
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic_data['Age'], titanic_data['Fare'], color='green', marker='o')

# Add labels and title
plt.xlabel('Age of Passenger')
plt.ylabel('Fare Amount')
plt.title('Scatter Plot of Age vs Fare')

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

Bubble Chart

```python
# Define color map for passenger class
color_map = {1: 'blue', 2: 'green', 3: 'red'}
colors = titanic_data['Pclass'].map(color_map)

# Create bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(
    titanic_data['Age'],
    titanic_data['Fare'],
    s=titanic_data['SibSp'] * 20,  # Scale sibsp for better visualization
    c=colors,
    alpha=0.5
)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Bubble Chart of Age vs Fare on Titanic')
plt.colorbar(label='Passenger Class', ticks=[1, 2, 3], format='%d')

# Add grid
plt.grid(True)
```

Histogram

```python
# Drop rows with missing values in 'age' column
titanic_data.dropna(subset=['Age'], inplace=True)

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(titanic_data['Age'], bins=30, edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Passenger Ages on Titanic')

# Show plot
plt.show()
```

Pie Chart

```python
class_counts = titanic_data['Pclass'].value_counts()

# Define labels and sizes
labels = class_counts.index.map({1: 'First Class', 2: 'Second Class', 3: 'Third Class'})
sizes = class_counts.values

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99'])

# Add title
plt.title('Distribution of Passenger Classes on Titanic')

# Show plot
plt.show()
```
